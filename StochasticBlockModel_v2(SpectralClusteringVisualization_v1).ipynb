```python
import plotly.graph_objects as go
import random
import numpy as np
from numpy import linalg as LA
import sklearn.manifold as sk
```

```python
n, k, p, q , MinOutsideWeight, MaxOutsideWeight, MinInsideWeight, MaxInsideWeight = input().split()
#Clusters Size
C = np.zeros(int(k))
#Adjacency Matrix
M = np.zeros((int(n),int(n)))
#Degree Matrix
D = np.zeros((int(n),int(n)))
#Laplacian
L = np.zeros((int(n),int(n)))

color_Map = []
```

```python
t = 1
for i in range(int(k)):
    C[i] = int(input())
    a = int(random.random() * 255)
    b = int(random.random() * 255)
    c = int(random.random() * 255)
    rgb = (a, b, c)
    for j in range(int(C[i])):
        color_Map.append('#' + str('%02x%02x%02x' % rgb))
        t += 1
```

```python
t = 0
cnt = 0
for i in range(int(n)):
    if(i >= cnt + C[t]):
        cnt += C[t]
        t += 1
    for j in range(i):
        if(j >= cnt):
            x = random.random()
            if(x <= float(p)):
                y = random.random()
                y *= (int(MaxInsideWeight) - int(MinInsideWeight))
                y += int(MinInsideWeight)
                M[i][j] = M[j][i] = y
        else:
            x = random.random()
            if(x <= float(q)):
                y = random.random()
                y *= (int(MaxOutsideWeight) - int(MinOutsideWeight))
                y += int(MinOutsideWeight)
                M[i][j] = M[j][i] = y
```

```python
#M = [[0, 9, 0, 0], [9, 0, 1, 0], [0, 1, 0, 9], [0, 0, 9, 0]]
for i in range(int(n)):
    s = 0
    for j in range(int(n)):
        s += M[i][j]
    D[i][i] = s
```

```python
def normalize(V, N):
    for i in range(N):
        s = 0
        for j in range(3):
            s += V[i][j] ** 2
        s = np.sqrt(s)
        for j in range(3):
            V[i][j] /= s
            V[i][j] = float('%.8f'%(V[i][j]))
```

```python
for i in range(int(n)):
    for j in range(int(n)):
        L[i][j] = D[i][j] - M[i][j]
eVals, eVectors = LA.eig(np.array(L))
idx = eVals.argsort()[::]   
eVals = eVals[idx]
eVectors = eVectors[:,idx]
#normalize(eVectors, int(n))
```

```python
x = [0 for k in range(int(n))] 
x = [eVectors[k][1] for k in range(int(n))] 
y = [eVectors[k][2] for k in range(int(n))] 
z = [eVectors[k][3] for k in range(int(n))] 

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, 2 * np.pi, 100)

xB = 0.75 * np.outer(np.cos(u), np.sin(v))
yB = 0.75 * np.outer(np.sin(u), np.sin(v))
zB = 0.75 * np.outer(np.ones(np.size(u)), np.cos(v))

fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size= 15,
        color= color_Map,             
        colorscale='Viridis',   
        opacity=0.8
    )
), go.Surface(x = xB, y = yB, z = zB, opacity = 0.1)])

fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```
